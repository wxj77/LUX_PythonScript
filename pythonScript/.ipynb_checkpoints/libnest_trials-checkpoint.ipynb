```python
import sys
import os
sys.path.insert (2, '/global/homes/s/salsum/luxsalsum/libNEST/libNESTShaun')
#sys.path.insert(2,'/global/projecta/projectdirs/lux/Tools/fastNEST/libNEST/')
import libNEST
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import LogNorm
import matplotlib.pylab as pylab
pylab.rcParams["figure.figsize"]=15,8
```

```python
minDrift = 30
maxDrift = 105
field = 283
```

```python
g1 = .105
ee = .67
e_life = 1021
s2_thresh = 150
g1_gas = .089
ff = .67
```

```python
NEST = libNEST.NEST(1,1,1.,2.88,-1)
myDetector = libNEST.Detector()
```

```python
myDetector.LUXRun04PlanBSettings()
myDetector.g1 = g1
myDetector.ee = ee
myDetector.e_life = e_life
myDetector.dt_min
myDetector.g1_gas = g1_gas
myDetector.ff = ff

NEST.SetDetectorParameters(myDetector)
NEST.SetElectricField(field)
NEST.SetSingleElectron(False)
```

```python
S1csl = []
S2csl = []
for aa in range(100000):
    energy = libNEST.TritiumEnergy()
    NEST.SetEnergy(energy)
    NEST.DetectorResponse()
    for ee in range(0,)NEST.SetElectricField()
    if (NEST.GetS1c() > 0) and (NEST.GetS2c() > 0):
        if (((np.log10(NEST.GetS1c())+0.7*np.log10(NEST.GetS2c()))<4.8)
           and ((np.log10(NEST.GetS1c())+0.7*np.log10(NEST.GetS2c()))>2.3)
           and ((np.log10(NEST.GetS1c())-2*np.log10(NEST.GetS2c()))<-5)
           and ((np.log10(NEST.GetS1c())-2*np.log10(NEST.GetS2c()))>-6.5)):
            S1csl.append(NEST.GetS1c())
            S2csl.append(NEST.GetS2c())
```

```python
S1cs = np.array(S1csl)
S2cs = np.array(S2csl)
```

```python
ax = plt.subplot()
plt.hist2d(np.log10(S1cs), np.log10(S2cs), range=[[1,2.5],[2.8,4.5]], bins=(100,100), norm=LogNorm())
plt.xlabel("Log S1c", fontsize=18)
plt.ylabel("Log S2c", fontsize=18)
plt.title("libNEST Tritium Sep 2014 Top Slice", fontsize=20)
```

```python
NESTdata = [[],[]]
for aa in range(9):
    s1bins1 = []
    s1bins2 = []
    for bb in range(len(S1cs)):
        if (aa+0.5 < S1cs[bb]) & (S1cs[bb] < aa+1.5):
            s1bins1.append(S1cs[bb])
            s1bins2.append(S2cs[bb])
    NESTdata[0].append(s1bins1)
    NESTdata[1].append(s1bins2)
```

```python
fig1 = plt.figure(1)
plt.subplots_adjust(top=1.15)
ax = fig1.add_subplot(111)
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top='off', bottom='off', left='off', right='off')
ax.set_xlabel("Log S2", fontsize=30)
ax.set_ylabel("Counts", fontsize=30)
ax.set_title("Top Slice", fontsize=30, y=1.08)
for aa in range(9):
    fig1.add_subplot(3,3,aa+1)
    plt.hist(np.log10(NESTdata[1][aa]), range=[2.5,4.5], bins=30, normed=True, histtype="step")
    plt.title(str(aa+.5)+" - "+str(aa+1.5)+" S1", fontsize=20)
```
